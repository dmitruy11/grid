## создать модель антенной решётки при помощи заданной функции с возможностью выбора параметров

```python

import numpy as np
import matplotlib.pyplot as plt

# Параметры антенной решетки
elements = 10             # Число элементов решетки
spacing = 0.5             # Расстояние между элементами (в длинах волн)
angle = np.linspace(0, 180, 181)  # Угол сканирования (в градусах)

# Создание модели антенной решетки
def array_pattern(n, d, theta):
    """
    Функция для расчета диаграммы направленности антенной решетки.
    Аргументы:
        n - число элементов решетки
        d - расстояние между элементами (в длинах волн)
        theta - угол сканирования (в градусах)
    Возвращает:
        AF - диаграмма направленности
    """
    k = 2 * np.pi / 1  # волновое число
    phi = k * d * np.cos(np.deg2rad(theta))
    AF = np.abs(np.sin(n * phi / 2) / (n * np.sin(phi / 2)))
    return AF

# Построение диаграммы направленности
fig, ax = plt.subplots(figsize=(10, 6))
AF = array_pattern(elements, spacing, angle)
ax.plot(angle, AF)
ax.grid()
ax.set_xlabel('Угол, градусы')
ax.set_ylabel('Диаграмма направленности')
ax.set_title('Диаграмма направленности антенной решетки')

# Возможность изменения параметров
n = int(input('Введите число элементов решетки (от 2 до 20): '))
d = float(input('Введите расстояние между элементами (в длинах волн, от 0.25 до 1): '))

fig, ax = plt.subplots(figsize=(10, 6))
AF = array_pattern(n, d, angle)
ax.plot(angle, AF)
ax.grid()
ax.set_xlabel('Угол, градусы')
ax.set_ylabel('Диаграмма направленности')
ax.set_title('Диаграмма направленности антенной решетки')

plt.show()
```
